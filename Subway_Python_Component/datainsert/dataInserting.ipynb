# Getting a sorted list of all the stations

```python
ogset = set()
for i in range(1,11):
  for r in df[str(i)]:
    if type(r) == float:
      break
    ogset.add(r)
print(ogset)
```

```python
namelist = list(ogset)
```

```python
namelist = sorted(namelist)
namelist
```

```python
for station in namelist:
  print(f'<option value="{station}">{station}</option>')
```

# Getting the ArrayLists names for the routesToStation

```python
for i in range(1,11):
  current = df[str(i)]
  res = []
  for station in current:
    if type(station) == float:
      break
    res.append(station)
  new_df[i] = res
print(new_df)
```

```python
for i in range(1,11):
  java_arraylist = 'ArrayList<String> names' + str(i) + '= new ArrayList<>(Arrays.asList(' + \
    ', '.join(f'"{name}"' for name in new_df[i]) + '));'
  print(java_arraylist)
```

# Obtaining stationToRoute

```python
from collections import defaultdict
name_df = defaultdict(list)
```

```python
for i in range(1,11):
  current = df[str(i)]
  for station in current:
    if type(station) == float:
      break
    name_df[station].append(i)
print(new_df)
```

```python
name_df
```

```python
len(name_df)
```

```python
i = 0
for u in name_df.keys():
  java_arraylist = 'ArrayList<Integer> stationToRoutes' + str(i) +'= new ArrayList<>(Arrays.asList(' + \
    ', '.join(f'{id}' for id in name_df[u]) + '));'
  putting = f'stationToRoutes.put("{u}",stationToRoutes{i});'
  i+=1
  print(java_arraylist)
  print(putting)
```
